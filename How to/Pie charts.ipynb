# --Matplolib Pie Charts

Ref: http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.pie

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

## Create dataframe

```python
raw_data = {'officer_name': ['Jake', 'Amy', 'Terry', 'Charles', 'Rosa'],
        'jan_arrests': [10, 13, 31, 2, 3],
        'feb_arrests': [13, 9, 10, 2, 7],
        'march_arrests': [21, 18, 22, 3, 5]}
df = pd.DataFrame(raw_data, columns = ['officer_name', 'jan_arrests', 'feb_arrests', 'march_arrests'])
df
```

```python
# Create/add a column with the total arrests for each officer
df['total_arrests'] = df['jan_arrests'] + df['feb_arrests'] + df['march_arrests']
df
```

## Pie chart with slice names close to slices

```python
# Define pie chart options
colors = ["#96ceb4", "#ffeead", "#88d8b0", "#ff6f69", "#ffcc5c"]
explode =  [0.15, 0, 0, 0, 0] #  Explode 1st slice
autopct = '%1.1f%%' # Display percentage within the pie slices: '%1.1f%%' for 1 decimal precision

# Create pie chart with slice names close to slice
plt.pie(
    # to show the total arrests
    df['total_arrests'],
    # with the labels officer names
    labels=df['officer_name'],
    # with shadows
    shadow=True,
    # with colors
    colors=colors,
    # with one slide exploded out
    explode=explode,
    # with the start angle at 0 degree
    startangle=0,
    # with the percentages on pie
    autopct=autopct,
    )

# View the plot drop above
plt.axis('equal')

# View the plot
plt.tight_layout()
plt.show()
```

## Pie chart with slice name in legend

```python
labels=['Jake', 'Amy', 'Terry', 'Charles', 'Rosa']

# Create pie chart with slice name in legend
slices = plt.pie(
    # to show the total arrests
    df['total_arrests'],
    # with shadows
    shadow=True,
    # with colors
    colors=colors,
    # with one slide exploded out
    explode=explode,
    # with the start angle at 0 degree
    startangle=0,
    # with the percentages on pie
    autopct=autopct,
    )
# View the plot drop above
plt.axis('equal')
plt.legend(labels, loc="best")
# View the plot
plt.tight_layout()
plt.show()
```
