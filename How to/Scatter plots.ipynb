# Scatter plots

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Scatter plots with varying dot sizes

```python
# Create a dummy dataframe with 2 columns X and Y
df = pd.DataFrame(np.random.randn(50, 2), columns=['X','Y'])
# Derive third column Z
df['Z'] = abs(df.X * df.Y) * 1000
df.head()
```

```python
# Scatter plot
plt.scatter(df.X, df.Y, s=df.Z)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Example of scatter plot with varying dot sizes')
```

## Scatter plots with coloured dots

```python
# Define 4 classes
class1 = df[(df.X > 0) & (df.Y > 0)]
class2 = df[(df.X > 0) & (df.Y < 0)]
class3 = df[(df.X < 0) & (df.Y < 0)]
class4 = df[(df.X < 0) & (df.Y > 0)]

# Scatter plot
plt.scatter(class1.X, class1.Y, s=class1.Z, c='lightblue')
plt.scatter(class2.X, class2.Y, s=class2.Z, c='orange')
plt.scatter(class3.X, class3.Y, s=class3.Z, c='lightgreen')
plt.scatter(class4.X, class4.Y, s=class4.Z, c='pink')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Example of scatter plot with coloured dots')
```
