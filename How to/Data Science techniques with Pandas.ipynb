# Data Science techniques with Pandas

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../DataScience/datasets/titanic/train.csv")
df.head()
```

## One Hot Encoding of categorical datas --> pd.get_dummies()

```python
sex = ['male', 'female']
pd.get_dummies(sex)
```

```python
# If a catgorical value can only take 2 values, e.g. sex: male or female, we can use the option drop_first=True
df1 = pd.get_dummies(df, columns=['Sex'], drop_first=True)
df1.head()
```

```python
# Other option: Create and merge dummies in the same data frame
df2 = df.join(pd.get_dummies(df['Sex'], drop_first=True))
df2.head()
```

```python
df['Embarked'].unique()
```

```python
# Check the NaN values
df[df['Embarked'].isnull()]
```

```python
# As NaN values are possible for the 'Embarked' column use the 'dummy_na' option to consider them
df3 = pd.get_dummies(df, columns=['Sex', 'Embarked'], dummy_na=True, drop_first=True)
df3.head()
```
